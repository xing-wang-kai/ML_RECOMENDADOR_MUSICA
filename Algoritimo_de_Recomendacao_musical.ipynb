```python
data_by_genrer_v1 = data_by_genrer.drop('genres', axis=1).copy()
data_by_genrer_v1.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
pca_pipeline = Pipeline([('saler', StandardScaler()), ('PCA', PCA(n_components=2, random_state=SEED))])
```

```python
genre_embedding_pca = pca_pipeline.fit_transform(data_by_genrer_v1)
projection = pd.DataFrame(columns=['x', 'y'], data=genre_embedding_pca)
```

```python
projection.head()
```

```python
from sklearn.cluster import KMeans

kmeans_pca = KMeans(n_clusters=5, verbose=True, random_state=SEED)
kmeans_pca.fit(projection)
data_by_genrer['cluster_pca'] = kmeans_pca.predict(projection)
projection['cluster_pca'] = kmeans_pca.predict(projection)
```

```python
projection.head()
```

```python
projection['genrer'] = data_by_genrer['genres']
projection.head()
```

```python
figure = px.scatter(
    projection, x = 'x', y='y', color='cluster_pca', hover_data=['x', 'y', 'genrer']
)

figure.show()
```

```python
# ESTE METODO E APLICADO A PIPLINES
# RETORNA CLASS COMO GETDUMMYS porém com mais dados.

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(dtype=int)
columns_ohe = ohe.fit_transform(data_total[['artists']]).toarray()

columns_ohe
```

```python
data_total_v2 = data_total.drop('artists', axis=1).copy()

data_total_dummies = pd.concat([data_total_v2, pd.DataFrame(columns_ohe, columns=ohe.get_feature_names_out(['artists']))], axis=1)
data_total_dummies
```

```python
data_total_dummies_v1 = data_total_dummies.drop(['id', 'name', 'artists_song'], axis=1).copy()
```

```python
data_total_dummies_v1.columns
```

```python
pca_pipeline = Pipeline([('saler', StandardScaler()), ('PCA', PCA(n_components=0.7, random_state=SEED))])

music_embedding_pca = pca_pipeline.fit_transform(data_total_dummies_v1)
projection_m = pd.DataFrame(  data=music_embedding_pca)
```

```python
kmeans_pca = KMeans(n_clusters=50, verbose=True, random_state=SEED)
kmeans_pca.fit(projection_m)
data_total_dummies_v1['cluster_pca'] = kmeans_pca.predict(projection_m)
projection_m['cluster_pca'] = kmeans_pca.predict(projection_m)
```

```python
nome_music = 'Ed Sheeran - Shape of You'
```

```python
from sklearn.metrics.pairwise import euclidean_distances

cluster = list(projection_m[projection_m['song'] == nome_music]['cluster_pca'])[0]
music_recomendadas = projection_m[projection_m['cluster_pca'] == cluster][[0, 1, 'song']]
x_music = list(projection_m[projection_m['song']== nome_music][0])[0]
y_music = list(projection_m[projection_m['song']== nome_music][1])[0]

#Distancia euclidiana
# Separa a distancia de valores das columas 0 e 1 da musica selecionada que estão mais prox da columas 0 e 1 das musicas recomendadas

distance = euclidean_distances(music_recomendadas[[0,1]], [[x_music, y_music]])

# pega o valor 0 e 1 para eixo x e y da ROW onde o song é igual a musica recomendada.
music_recomendadas['id'] = data_total['id']
music_recomendadas['distance'] = distance


# Organiza a tabela do menor ao maior valor da distancia recomendada.
recomendadas = music_recomendadas.sort_values('distance').head(10)
recomendadas
```
